```python
import pandas as pd
import numpy as np
levels = ['C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 
          'T1', 'T2', 'T3', 'T4','T5', 'T6', 'T7', 'T8', 'T9', 'T10', 'T11', 'T12', 
          'L1', 'L2','L3', 'L4', 'L5']
```

```python
df_raw = pd.read_excel("Supp_values.xlsx", header=[0,1], index_col = 0, skipfooter=1, sheet_name="lenght")
df_raw.head()
```

```python
df_mm = df_raw.copy()*10
df_mm["Width across columns (mm)"] = df_raw["Width across columns (mm)"]
df_mm.columns = df_mm.columns.set_levels(df_mm.columns.levels[0].str.replace("cm","mm"), level=0)
df_mm.head()
```

```python
df_mm_spl = df_mm.copy()
df_mm_spl.columns = ['_'.join(col) for col in df_mm_spl.columns.values]
df_mm_spl.head()
```

```python
mesure_name = ["DREZ length","Seg.L at dorsal CE","Seg.L at BE","Inf.Art facet to CR distance","IF to RR distance","IF to CR distance", "Dorsal width"]
new_df = {"Segment":[], "MEAN":[], "STD":[], "Mesure":[]}
for idx in df_mm_spl.index:
    line = df_mm_spl.loc[idx]
    new_df["MEAN"].extend(line.values[::2])
    new_df["STD"].extend(line.values[1::2])
    new_df["Mesure"].extend(mesure_name)
    new_df["Segment"].extend([idx]*len(mesure_name))
df_concat = pd.DataFrame(new_df)
df_concat.head()
```

```python
df_concat.to_csv("mendez_df.csv", index = False)
```
